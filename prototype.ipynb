## FTR, HTR
홈팀이 이겼을때 더 가중치를 주기위해 홈팀이 이겼을때 1, 비겼을때 0, 졌을때 -1로 전처리

```python
ftr_mapping = {'H': 1, 'D': 0, 'A': -1}
df['FTR'] = df['FTR'].map(ftr_mapping).astype(float)
```

```python
htr_mapping = {'H': 1, 'D': 0, 'A': -1}
df['HTR'] = df['HTR'].map(htr_mapping).astype(float)
```
