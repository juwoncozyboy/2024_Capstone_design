```python
def home_point(result):
    if result == 'H':  # 홈팀 승
        return 3
    elif result == 'A':  # 원정팀 승
        return 0
    else:  # 무승부
        return 1

def away_point(results):
    if results == 'H': # 홈팀 승
        return 0 # 원정팀은 진 경우이므로 0
    elif results == 'A':
        return 3 # 원정팀이 이긴경우니까 3
    else:
        return 1
```

```python
def arsenal_points(row, i):
    if row['HomeTeam'] == 'Arsenal':
        point = home_point(row['FTR'])
    elif row['AwayTeam'] == 'Arsenal':
        point = away_point(row['FTR'])
    else:
        point = 0
    
    i['Arsenal'] += point
    return i['Arsenal']

i = {'Arsenal': 0}


big6['Arsenal_Points'] = big6.apply(lambda row: arsenal_points(row, i), axis=1)


```

## FTR, HTR
홈팀이 이겼을때 더 가중치를 주기위해 홈팀이 이겼을때 1, 비겼을때 0, 졌을때 -1로 전처리

```python
ftr_mapping = {'H': 1, 'D': 0, 'A': -1}
big6['FTR'] = big6['FTR'].map(ftr_mapping).astype(float)
```

```python
htr_mapping = {'H': 1, 'D': 0, 'A': -1}
big6['HTR'] = big6['HTR'].map(htr_mapping).astype(float)
```
